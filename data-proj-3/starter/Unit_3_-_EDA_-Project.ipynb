### Step 1 - Review Summary Statistics

- mean
> all values add together and then divided by the number of values (the average)
- median
> average of the middle value(s) in a data set (4,5,6,9,44) -> 6 (4,5,6,8,19,22) -> 7
- mode
> the most frequent item in a distribution (4, 5, 6, 6, 6, 7, 7, 8) -> 6... in a KDE mode is always the peak
- central tendency
> statistics that measure the average values of data sets (typically mean, median, mode) looking for central value
- distribution
> how values are distributed across a given field... shows if a value is common or uncommon
- variance
> shows variability or the size of the spread in your data set by taking the mean squared distance between each data point and the center of the distrubiton measured by the mean. 
- standard deviation
> square root of the variance measures how dispersed the data is in relation to the mean.. low standard deviation means data is close to mean high standard deviation means data is spread out

### Step 2 - Review Hypothesis Testing & P-Value

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

#### Fill in blanks with brief description of the following terms:

- **Hypothesis**
> A hypothesis is an assumption based on limited evidence that requires more investigation

- **Null Hypothesis**
> A statement that assumes there is no significant relationship between two or more variables.. ie years of work experiences will not affect the salary of managers

- **P-Value**
> The p-value tells us whether or not we can reject our null hypothesis. It is on a scale of 0 to 1. Typically if a p-value is equal to or below 0.05 (which means that there would be a 5% or less chance that the correlation is random) it is safe to reject the null hypothesis and dive into deeper research on the alternative hypothesis.

### What is the correlation between salary and education?

#### Your Hypothesis = salary will increase as education level increases so a positive correlation. 

```python
# calculate correlation and pvalue of salary and education with spearmanr method
```

```python
pearsonr(df_sn['salary'], df_sn['education'])
```

```python
spearmanr(df_sn['salary'], df_sn['education'])
```

[interpret result] we can see that correlation is positive and the p-value = 0... Shows that our alternative hypothesis is worth exploring further and null hypothesis should be rejected.

#### Are these results statistically significant? How do you know?

Yes they are statistically significant because in both the spearman and pearson methods the p-value is 0 and both have strong positive correlations(salary increases with education level increase). This means there is a 0 percent chance that positive correlation is due to randomness and we should reject the null hypothesis and explore further.

#### Fill in the blanks with a brief description of each plot type 
  
- pairplots
> A pairplot is going to show pairwise scatter plots for each combination of variables in a data set. Because each variable is plotted against every other variable we end up with a grid visualization. 
- paired scatter plots
> Similiar to a pairplot in that we are shown a data visualization where all numeric variables are compared via a scatter plot. This allows us to see relationships and is a great tool for quick EDA. 
- correlation matrix
> Rather than create a visualizsation, a correlation matrix will show positive to negative correlation between variables via a df like chart. With -1 being associated with negative correlation, 0 with no correlation, and 1 with positive correlation, we can quickly see relationships between numeric variables.
- heatmap
> A heat plot shows the correlation between all variables in a given data frame by making a grid where each variable in on both the x and y axis. The heatmap typically uses red to show positive correlation, white to show no correlation, and blue to show negative correlation.
